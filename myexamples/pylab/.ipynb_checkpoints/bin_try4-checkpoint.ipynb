### Binary simulation output plotting

```python
# plot some stuff
def mkplot(tarr,aaarr,eearr,iiarr,lnarr,ararr,maarr,om1,om2,\
        obliquity_deg1,obliquity_deg2,meanmotion,lib_angle,froot,dn,ofilename):
    # dn is to plot every other point
    nvpanels = 5
    f,axarr =  plt.subplots(nvpanels,1, dpi=150, figsize=(5,6), sharex=True)
    plt.autoscale(enable=True, axis='x', tight=True)
    plt.subplots_adjust(left=0.2, right=0.99, top=0.99, bottom=0.10, \
        wspace=0.22, hspace=0.0)
    
    Porb_init = 2*np.pi/meanmotion[0]
    tarr /= Porb_init
    
    ###########################
    il = 0;   # top
    smax = np.amax([np.amax(om1),np.amax(om2)])
    smin = np.amin([np.amin(om1),np.amin(om2)])
    axarr[il].set_ylim(smin,smax)
    axarr[il].plot(tarr[::dn],om1[::dn],'.',color='black',\
                   ms=2,label='prim')
    axarr[il].plot(tarr[::dn],om2[::dn],'.',color='blue',\
                   ms=2,label='sec')
    mm_smo = meanmotion
    if (len(tarr)>100):
        mm_smo = savgol_filter(meanmotion, 21, 2, mode='nearest')
    axarr[il].plot(tarr[::dn],mm_smo[::dn]*1,':',color='brown',lw=2,alpha=0.5)
    for j in range(2,7):
        axarr[il].plot(tarr[::dn],mm_smo[::dn]*j,':',color='brown',lw=1,alpha=0.5)
    #for j in range(0,7):
    #    axarr[il].plot(tarr[::dn],mm_smo[::dn]*(j+0.5),':',color='purple',lw=1,alpha=0.5)
        
    axarr[il].set_ylabel('spin')
    axarr[il].legend(borderpad=0.1,labelspacing=0.1,handlelength=0.5,handletextpad=0.1,\
                    markerscale=2,loc='upper left')
    
    ##########################
    il = 1;
    axarr[il].plot(tarr[::dn],obliquity_deg1[::dn],'.',color='red',\
                   ms=2,label='prim')
    axarr[il].plot(tarr[::dn],obliquity_deg2[::dn],'.',color='darkorange',\
                   ms=2,label='sec')
    axarr[il].set_ylabel('obliquity \n (deg)')
    axarr[il].legend(borderpad=0.1,labelspacing=0.1,handlelength=0.5,handletextpad=0.1,\
                    markerscale=2)
    
    
    ##########################
    il = 2;
    lib_a = lib_angle%np.pi
    ii = (lib_a > np.pi/2)
    lib_a[ii] -= np.pi
    lib_a *= 180.0/np.pi
    axarr[il].plot(tarr[::dn],lib_a[::dn],'.',color='purple',ms=2)
    axarr[il].set_ylabel('libration \n angle (deg)')
    #axarr[il].set_ylim(bottom=2.5,top=5)
    
    ##########################
    il = 3;
    amax = max(aaarr)
    amin = min(aaarr)
    if (amax > 5):
        axarr[il].set_ylim(max(amin,0),5)
    axarr[il].plot(tarr[::dn],aaarr[::dn],'.',color='green',ms=2)
    axarr[il].set_ylabel('semi-major \n axis ' + r'$a_B/R_p$')
    #axarr[il].set_ylim(bottom=2.5,top=5)
    
    ########################## 
    il = 4;
    emax = max(eearr)
    emin = min(eearr)
    if (emax > 1):
        axarr[il].set_ylim(0,1)
    axarr[il].plot(tarr[::dn],eearr[::dn],'.',color='dodgerblue',ms=2)
    axarr[il].set_ylabel('eccentricity')
    il = nvpanels-1;
    axarr[il].set_xlabel('time (orbital periods)')
    if (len(ofilename)>3):
        plt.savefig(ofilename)
```

$A = \frac{M}{5} (b^2 + c^2) $

$B = \frac{M}{5} (a^2 + c^2) $

$C = \frac{M}{5} (a^2 + b^2) = \frac{M}{5} a^2(1 + \frac{b}{a} )^2)) $

$A<B<C$

$A+B+C = \frac{2M}{5}(a^2 + b^2 + c^2) $ 

$V = \frac{4}{3} \pi abc = \frac{4}{3} \pi a^3 \frac{b}{a} \frac{c}{a}  = \frac{4}{3} \pi R^3$ 

If $R=1$ then 
$a = \left[ \frac{b}{a} \frac{c}{a} \right]^{-\frac{1}{3}} $

$ A + B - C = \frac{2}{5} Mc^2$

$ A + C - B = \frac{2}{5} Mb^2$

$ -A + B+ C = \frac{2}{5} Ma^2$


```python
def print_axis_ratios(I1,I2,I3):
    a_p = np.sqrt( (5.0/2.0)*(I1 + I2 - I3) )
    c_p = np.sqrt( (5.0/2.0)*(I2 + I3 - I1) )
    b_p = np.sqrt( (5.0/2.0)*(I3 + I1 - I2) )
    print('a,b,c= {:.3f} {:.3f} {:.3f}'.format(a_p,b_p,c_p))
    print('b/a = {:.3f}, c/a = {:.3f}'.format(b_p/a_p, c_p/a_p))  # real axis ratios of primary
```

```python
# from a1_run.txt
I1_p,I2_p,I3_p = 0.480784, 0.449625, 0.440806
I1_s,I2_s,I3_s = 0.000316, 0.000276, 0.000208
```

```python
print('primary:')
print_axis_ratios(I1_p,I2_p,I3_p)
print('secondary:')
print_axis_ratios(I1_s,I2_s,I3_s)
```

```python
# plot some stuff
def mkplot2(tarr,aaarr,eearr,iiarr,lnarr,ararr,maarr,om1,om2,\
        obliquity_deg1,obliquity_deg2,meanmotion,lib_angle,froot,dn,pmax,ofilename):
    # dn is to plot every other point
    nvpanels = 5
    f,axarr =  plt.subplots(nvpanels,1, dpi=150, figsize=(5,6), sharex=True)
    plt.autoscale(enable=True, axis='x', tight=True)
    plt.subplots_adjust(left=0.2, right=0.99, top=0.99, bottom=0.10, \
        wspace=0.22, hspace=0.0)
    
    Porb_init = 2*np.pi/meanmotion[0]
    tarr /= Porb_init
    
    
    if (pmax>1):
        axarr[0].set_xlim([0,pmax])
    
    ###########################
    il = 0;   # top
    smax = np.amax([np.amax(om1),np.amax(om2)])
    smin = np.amin([np.amin(om1),np.amin(om2)])
    axarr[il].set_ylim(smin,smax)
    axarr[il].plot(tarr[::dn],om1[::dn],'.',color='black',\
                   ms=2,label='prim')
    axarr[il].plot(tarr[::dn],om2[::dn],'.',color='blue',\
                   ms=2,label='sec')
    mm_smo = meanmotion
    if (len(tarr)>100):
        mm_smo = savgol_filter(meanmotion, 21, 2, mode='nearest')
    axarr[il].plot(tarr[::dn],mm_smo[::dn]*1,':',color='skyblue',lw=2,alpha=0.7)
    for j in range(2,7):
        axarr[il].plot(tarr[::dn],mm_smo[::dn]*j,':',color='skyblue',lw=1,alpha=0.5)
    #for j in range(0,7):
    #    axarr[il].plot(tarr[::dn],mm_smo[::dn]*(j+0.5),':',color='purple',lw=1,alpha=0.5)
        
    axarr[il].set_ylabel('spin')
    axarr[il].legend(borderpad=0.1,labelspacing=0.1,handlelength=0.5,handletextpad=0.1,\
                    markerscale=2)
    
    t1=900; s1 = 0.47; s2=0.30
    axarr[il].plot([0,t1],[s1,s2],':',color='cyan', alpha=0.7)
    dspin_dt = -(s2-s1)/(t1 *Porb_init)
    
    print('spin down rate {:.2e} nbody units'.format(dspin_dt))
    mm_smo_f = mm_smo[::dn]
    ss  = om2[::dn] - mm_smo[::dn]
    wid_len=11
    ss_smo = savgol_filter(ss,wid_len,3,mode='constant')
    lss_smo = np.log(np.abs(ss_smo))
    tt = tarr[::dn]
    ii = tt<1100
    tt = tt[ii]*Porb_init
    lss_smo = lss_smo[ii]
    ssp = polyfit(tt,lss_smo,1)
    ssfun = np.poly1d(ssp)
    mm_smo_f = mm_smo[::dn]
    axarr[il].plot(tt/Porb_init,np.exp(ssfun(tt))+ mm_smo_f[ii],':',color='yellow')
    
    ##########################
    il = 1;
    axarr[il].plot(tarr[::dn],obliquity_deg1[::dn],'.',color='red',\
                   ms=2,label='prim')
    axarr[il].plot(tarr[::dn],obliquity_deg2[::dn],'.',color='darkorange',\
                   ms=2,label='sec')
    wid_len=41
    tt = tarr[::dn]
    oo = savgol_filter(obliquity_deg2[::dn],wid_len,3,mode='constant')
    loo = np.log(np.abs(oo))
    ii = tt>980
    tt = tt[ii]*Porb_init
    oo = oo[ii]
    loo = loo[ii]
    pp = np.polyfit(tt,oo,1)
    pfun = np.poly1d(pp)
    ppl = np.polyfit(tt,loo,1)
    pfunl = np.poly1d(ppl)
    print('polyfit ', pp)
    print('polyfitl ', ppl)
    dodt = -pp[0]*np.pi/180   #  in nbody units rad/time^2
    o_damp_rate = -ppl[0] # epsilon = epsilon_0 e^(-damprate*t) nbody units
    print('obliquity damping rate from log, nbody units  {:.3e}'.format(o_damp_rate))
    print ('obliquity damping rate (rad/nbody units) = {:.2e}'.format(dodt))
    print('ratio dodt/dspindt = {:.3f}'.format(dodt/dspin_dt))
    print('ratio dspindt/dodt = {:.3f}'.format(dspin_dt/dodt))
    dratio =o_damp_rate*meanmotion[0]/dspin_dt
    print('ratio damprate n_B /dspindt = {:.3f}'.format(dratio))
    print('ratio dspindt/(damprate n_B) = {:.3f}'.format(1/dratio))
    print('ratio dspindt/(damprate om_Br) = {:.3f}'.format(dspin_dt/o_damp_rate))

    #axarr[il].plot(tt,oo,'-',color='gray')
    axarr[il].plot(tt/Porb_init,pfun(tt),':',color='black')
    axarr[il].plot(tt/Porb_init,np.exp(pfunl(tt)),':',color='green')
    
    axarr[il].set_ylabel('obliquity \n (deg)')
    axarr[il].legend(borderpad=0.1,labelspacing=0.1,handlelength=0.5,handletextpad=0.1,\
                    markerscale=2)
    
    
    ##########################
    il = 2;
    lib_a = lib_angle%np.pi
    ii = (lib_a > np.pi/2)
    lib_a[ii] -= np.pi
    lib_a *= 180.0/np.pi
    axarr[il].plot(tarr[::dn],lib_a[::dn],'.',color='purple',ms=2)
    axarr[il].set_ylabel('libration \n angle (deg)')
    #axarr[il].set_ylim(bottom=2.5,top=5)
    
    ##########################
    il = 3;
    amax = max(aaarr)
    amin = min(aaarr)
    if (amax > 5):
        axarr[il].set_ylim(max(amin,0),5)
    axarr[il].plot(tarr[::dn],aaarr[::dn],'.',color='green',ms=2)
    axarr[il].set_ylabel('semi-major \n axis ' + r'$a_B/R_p$')
    #axarr[il].set_ylim(bottom=2.5,top=5)
    
    ########################## 
    il = 4;
    emax = max(eearr)
    emin = min(eearr)
    if (emax > 1):
        axarr[il].set_ylim(0,1)
    axarr[il].plot(tarr[::dn],eearr[::dn],'.',color='dodgerblue',ms=2)
    axarr[il].set_ylabel('eccentricity')
    il = nvpanels-1;
    axarr[il].set_xlabel('time (orbital periods)')
    if (len(ofilename)>3):
        plt.savefig(ofilename)
```

```python
# read in orbital elements, spins for both resolved bodies
froot = 'b1'  # tides alone trying to let secondary tidally lock
tarr,aaarr,eearr,iiarr,lnarr,ararr,maarr,om1,om2,\
        obliquity_deg1,obliquity_deg2,meanmotion,lib_angle = read_two_bodies(simdir,froot,GM)


# make a plot!
ofilename='b1.png'
pmax = 0
mkplot2(tarr,aaarr,eearr,iiarr,lnarr,ararr,maarr,om1,om2,\
        obliquity_deg1,obliquity_deg2,meanmotion,lib_angle,froot,10,pmax,ofilename)
```
